# Dátové štruktúry - LIST (zoznam != pole) !!!

Dátová štruktúra je spôsob, akým sú údaje uložené a organizované v pamäti počítača, aby sa s nimi dalo efektívne pracovať (ukladať ich, vyhľadávať, meniť, mazať).

# Dátová štruktúra POLE
Pole je dátová štruktúra, pri ktorej sa v pamäti počítača vyhradí súvislý pamäťový blok o zvolenej veľkosti (napr. n).
Do tohto bloku je možné ukladať prvky rovnakého typu, pričom k jednotlivým prvkom sa pristupuje pomocou indexu.

Základné vlastnosti poľa:
  - prvky majú rovnaký dátový typ
  - majú pevne danú veľkosť (v mnohých jazykoch)
  - k prvkom sa pristupuje pomocou indexu
  - indexovanie zvyčajne začína od 0
  - prvky sú v pamäti reťazené sekvenčne (nasledujú za sebou)

Príklad: Majme pole s názvom 'pole' o veľkosti n (n = 4) buniek a v bunkách ukladáme prvky '5', '8', '2', '9' : 
  - index:   0   1   2   3
  - pole:   [5,  8,  2,  9]

K jednotlivým bunkám pristupujeme pomocou indexu (!!! indexovanie poľa začína od '0' a končí hodnotou 'n-1' !!!):
  - Pre prístup k uloženému prvku v bunke používame zápis: 'pole[2]'
  - Pre ukladanie do bunky používame zápis: 'pole[1] = 4' 
(pole po zmene => [5,  4,  2,  9])

!!! Pozor nie je povolené prekračovať rozsah poľa, môžeme si prepísať dáta v pamäti, alebo čítať dáta ktoré neexistujú/nemajú zmysel pre náš program. 
Prekročenie indexov poľa môže viesť k nepredvídanému správaniu (nestabilný kód, problém zo spustením, nežiadúce správanie, atď). !!!

# Dátová štruktúra LIST
V pythone pole != pole, je to dátová štruktúra list(zoznam)!!!!
List nevyhradzuje v pamäti súvislý pamäťový blok, prvky nemusia byť uložené v pamúti za sebou.
Výhodou oproti poľu je, že list nepotrebuje okamžite zaberať v pamäti 'n' buniek, nepotrebujeme vedieť veľkosť dopredu.
List sa zväčšuje vplyvom narastajúceho počtu prvkov (list je dynamická dátová štruktúra).

List je:
  - usporiadaný
  - meniteľný (dá sa meniť)
  - môže obsahovať rôzne dátové typy
  - indexovanie začína od 0

Príklad vytvorenia zoznamu o veľkosti n (n=4) prvkov:
  - cisla = [1, 2, 3, 4]
  - cisla = [1, "ahoj", 3.14, True]

Príklad pridania nového prvku do zoznamu:
  - cisla.append(5)

Pridaním nového prvku do zoznamu sa zväčší veľkosť zoznamu na 5 prvkov, počet prvkov v zozname zistíme:
  - len(cisla)

Prvok zo zoznamu odstránime nasledovne (odstránením prvku sa zmenší počet prvkov v zozname):
  - cisla.remove("ahoj")
